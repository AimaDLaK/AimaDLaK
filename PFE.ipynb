```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
products_df = pd.read_csv('/content/products.csv')
user_interactions_df = pd.read_csv('/content/user_interactions.csv')
users_df = pd.read_csv('/content/users.csv')
# Print the head of each DataFrame

print("Head of products_df:")
print(products_df.head())
```

```python
print("\nHead of user_interactions_df:")
print(user_interactions_df.head())
```

```python
print("\nHead of users_df:")
print(users_df.head())
```

```python

# Print unique values and counts for specific columns
print("Unique product_id count:", products_df['product_id'].nunique())
print("Unique store_id count:", products_df['store_id'].nunique())

# Calculate and print average lengths
products_df['product_name_length'] = products_df['product_name'].str.len()
products_df['description_length'] = products_df['description'].str.len()
print("\nAverage product name length:", products_df['product_name_length'].mean())
print("Average description length:", products_df['description_length'].mean())

# Print unique values for boolean and categorical columns
print("\nUnique values for is_featured:", products_df['is_featured'].unique())
print("Unique values for is_on_sale:", products_df['is_on_sale'].unique())
print("Unique values for status:", products_df['status'].unique())
print("Unique values for has_image:", products_df['has_image'].unique())

# --- Distribution Plots ---

# Distribution of price
plt.figure(figsize=(10, 6))
sns.histplot(data=products_df, x='price', kde=True)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Distribution of regular_price
plt.figure(figsize=(10, 6))
sns.histplot(data=products_df, x='regular_price', kde=True)
plt.title('Distribution of Regular Price')
plt.xlabel('Regular Price')
plt.ylabel('Frequency')
plt.show()

# Distribution of sale_price
plt.figure(figsize=(10, 6))
sns.histplot(data=products_df, x='sale_price', kde=True)
plt.title('Distribution of Sale Price')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()

# Distribution of reviews_count
plt.figure(figsize=(10, 6))
sns.histplot(data=products_df, x='reviews_count', kde=True)
plt.title('Distribution of Reviews Count')
plt.xlabel('Reviews Count')
plt.ylabel('Frequency')
plt.show()

# Distribution of reviews_rating
plt.figure(figsize=(10, 6))
sns.histplot(data=products_df, x='reviews_rating', kde=True)
plt.title('Distribution of Reviews Rating')
plt.xlabel('Reviews Rating')
plt.ylabel('Frequency')
plt.show()


# Bar chart for num_images (as requested before)
plt.figure(figsize=(10, 6))
sns.countplot(data=products_df, x='num_images')
plt.title('Distribution of Number of Images per Product')
plt.xlabel('Number of Images')
plt.ylabel('Count')
plt.show()
```

```python

# Function to remove outliers using IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Distribution of price (outliers removed)
products_df_price_filtered = remove_outliers_iqr(products_df, 'price')
plt.figure(figsize=(10, 6))
sns.histplot(data=products_df_price_filtered, x='price', kde=True)
plt.title('Distribution of Price (Outliers Removed)')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Distribution of regular_price (outliers removed)
products_df_regular_price_filtered = remove_outliers_iqr(products_df, 'regular_price')
plt.figure(figsize=(10, 6))
sns.histplot(data=products_df_regular_price_filtered, x='regular_price', kde=True)
plt.title('Distribution of Regular Price (Outliers Removed)')
plt.xlabel('Regular Price')
plt.ylabel('Frequency')
plt.show()

# Distribution of sale_price (outliers removed)
products_df_sale_price_filtered = remove_outliers_iqr(products_df, 'sale_price')
plt.figure(figsize=(10, 6))
sns.histplot(data=products_df_sale_price_filtered, x='sale_price', kde=True)
plt.title('Distribution of Sale Price (Outliers Removed)')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()

# --- Bar Plots ---

# Distribution of reviews_count (Bar Plot)
plt.figure(figsize=(10, 6))
sns.countplot(data=products_df, x='reviews_count')
plt.title('Distribution of Reviews Count')
plt.xlabel('Reviews Count')
plt.ylabel('Count')
plt.show()

# Distribution of reviews_rating (Bar Plot)
plt.figure(figsize=(10, 6))
sns.countplot(data=products_df, x='reviews_rating')
plt.title('Distribution of Reviews Rating')
plt.xlabel('Reviews Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Drop specified columns
columns_to_drop = ['reviews_count', 'reviews_rating', 'is_featured', 'status']
products_df = products_df.drop(columns=columns_to_drop)

print("Columns remaining in products_df after dropping:", products_df.columns.tolist())




# --- Pie Plots ---

# Pie plot for is_on_sale
is_on_sale_counts = products_df['is_on_sale'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(is_on_sale_counts, labels=is_on_sale_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Products on Sale')
plt.show()

# Pie plot for has_image
has_image_counts = products_df['has_image'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(has_image_counts, labels=has_image_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Products with Images')
plt.show()
```

```python

# Drop rows where sale_price is 0 or price is 0
# Note: We are keeping products where regular_price is 0 if sale_price is not 0, or if both are non-zero.
# We only drop if sale_price is 0 OR price is 0.
products_df_cleaned = products_df[(products_df['sale_price'] != 0) & (products_df['price'] != 0)].copy()

# Drop products with 0 images
products_df_cleaned = products_df_cleaned[products_df_cleaned['num_images'] != 0].copy()

print(f"Shape of DataFrame after dropping products with sale_price=0 or price=0, and num_images=0: {products_df_cleaned.shape}")

# --- Replot Distributions of price and sale_price after cleaning (Outliers Handled) ---

# Function to remove outliers using IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Distribution of price (outliers removed) after cleaning
products_df_price_filtered_cleaned = remove_outliers_iqr(products_df_cleaned, 'price')
plt.figure(figsize=(10, 6))
sns.histplot(data=products_df_price_filtered_cleaned, x='price', kde=True)
plt.title('Distribution of Price after Cleaning (Outliers Removed)')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Distribution of sale_price (outliers removed) after cleaning
products_df_sale_price_filtered_cleaned = remove_outliers_iqr(products_df_cleaned, 'sale_price')
plt.figure(figsize=(10, 6))
sns.histplot(data=products_df_sale_price_filtered_cleaned, x='sale_price', kde=True)
plt.title('Distribution of Sale Price after Cleaning (Outliers Removed)')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()

# Distribution of regular_price (outliers removed) - replotting for completeness after cleaning
products_df_regular_price_filtered_cleaned = remove_outliers_iqr(products_df_cleaned, 'regular_price')
plt.figure(figsize=(10, 6))
sns.histplot(data=products_df_regular_price_filtered_cleaned, x='regular_price', kde=True)
plt.title('Distribution of Regular Price after Cleaning (Outliers Removed)')
plt.xlabel('Regular Price')
plt.ylabel('Frequency')
plt.show()
```

```python

# Print unique user counts
print("Unique user_id count:", users_df['user_id'].nunique())

# Calculate and print average length of bio
users_df['bio_length'] = users_df['bio'].str.len()
print("\nAverage bio length:", users_df['bio_length'].mean())

# --- Calculate Age from Birthdate ---
# Convert 'birthdate' to datetime objects
users_df['birthdate'] = pd.to_datetime(users_df['birthdate'])

# Calculate age based on current date
now = datetime.now()
users_df['age'] = users_df['birthdate'].apply(lambda x: now.year - x.year - ((now.month, now.day) < (x.month, x.day)))

print("\nAge column created.")
print("Sample ages:", users_df['age'].head())

# --- Pie Plots ---

# Pie plot for country_id
country_counts = users_df['country_id'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Countries')
plt.show()

# Pie plot for gender
gender_counts = users_df['gender'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Gender')
plt.show()

# Pie plot for language
language_counts = users_df['language'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(language_counts, labels=language_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Languages')
plt.show()


# Pie plot for is_public
is_public_counts = users_df['is_public'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(is_public_counts, labels=is_public_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Public Profile Status')
plt.show()

# Pie plot for is_profile_completed
is_profile_completed_counts = users_df['is_profile_completed'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(is_profile_completed_counts, labels=is_profile_completed_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Profile Completion Status')
plt.show()


# --- Bar Chart ---

# Bar chart for ages
# Using histplot as age is numerical and can have a wide range
plt.figure(figsize=(10, 6))
sns.histplot(data=users_df, x='age', bins=20, kde=True) # Use histplot for numerical age distribution
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()


# Unique values of preferences
print("\nUnique values for preferences:", users_df['preferences'].unique())
```
